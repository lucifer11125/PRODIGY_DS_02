```python
import pandas as pd
```

```python
#loading dataset
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.head(10)
```

```python
# Droping 'Cabin' column
df = df.drop(columns=['Cabin'])
```

```python
df.head(10)
```

```python
# Filling missing values in 'Age' with median
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df.head(10)
```

```python
# Filling missing values in 'Embarked' with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.head(10)
```

```python
# changing 'Sex' to binary values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df.head(10)
```

```python
# Convert 'Embarked' to one-hot encoding
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df.head(10)
```

```python
print(df.columns)

```

```python
df = df.drop(columns=['Name', 'Ticket', 'PassengerId'], errors='ignore')
```

```python
print(df.head(10))
```

```python
# EDA on the cleaned dataset
```

```python
#finding survial rate by gender 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=df, x='Sex', y='Survived')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Survival rate by passengers class
```

```python
sns.barplot(data=df, x='Pclass', y='Survived', errorbar=None)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Correlation Heatmap
```

```python
import numpy as np
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# relation between variables and patterns
```

```python
#Relationship between Gender and Survival
sns.countplot(data=df, x='Survived', hue='Sex', palette='viridis')
plt.title('Survival Count by Gender')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.show()
```

```python
#Pattern: Females tend to have a higher survival rate compared to males
```

```python
# Relationship between Passenger Class and Survival
sns.barplot(data=df, x='Pclass', y='Survived', hue='Pclass', legend=False, errorbar=None, palette='magma')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class (1 = First, 2 = Second, 3 = Third)')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Pattern: Passengers in higher classes (1st class) are more likely to survive compared to lower classes (3rd class)
```

```python
#Age Distribution of Survivors and Non-Survivors
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=20, palette='cool')
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.show()
```

```python
#Pattern: Younger passengers (e.g., children) show a higher survival rate compared to older ones.
```
